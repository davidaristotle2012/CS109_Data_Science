```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

x = np.linspace(0,10,30)
x
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y, label = "clean sine wave")
plt.plot(x,y + np.random.normal(size=30)*.3, label = "noisy sine wave")
plt.legend();
```
